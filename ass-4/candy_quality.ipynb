```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=pd.read_csv("/home/hmohan/Desktop/ass-4/candy-data.csv")
x=np.asmatrix(x)
x=np.array(x)
print(x.shape)
N,D=np.shape(x)
X=np.zeros((N,D-2))
print (X.shape)
#print("XXXXXXXXXXXX 1010101010",X[:,9])
X[:,0:10]=x[:,1:11]
Y=x[:,11]
print (X[0,9])

for i in range (len(X[:,9])):
    if X[i,9]<0.45:
        X[i,10]=0
    elif 0.45<=X[i,9]<0.86 :
        X[i,10]=1
    elif X[i,9] >=0.86:
        X[i,10]=-2

X=np.array(X,dtype=float)
Y=np.array(Y,dtype=float)
#X[:,10]=X[:,10]*100
#print("**********",X)
X[:,[9,10]] = X[:,[10,9]]
w=np.linalg.solve(np.dot(X[:,0:10].T,X[:,0:10]),np.dot(X[:,0:10].T,Y))
print("weights",w)
X[:,[9,10]] = X[:,[10,9]]
Y1=np.dot(X[:,:10],w)
print(Y1)
def sigmod(a):
    return (1/(1+np.exp(-a)))
Ys= sigmod(Y1)
print (Ys)
plt.scatter(np.sort(x[:,0].T),np.sort(Ys.T))
plt.plot(np.sort(x[:,0].T),np.sort(Ys.T))
plt.show()
#peanutyalmondy	nougat	crispedricewafer	hard	bar	pluribus	sugarpercent
'''
try:
    a= float(input("chocolate:"))
    b=float(input("fruty: "))
    c=float(input("caramil: "))
    d= float(input("peaNUT:"))
    E=float(input("nougat:"))
    f = float(input("wafer: "))
    g=float(input("hard :"))
    i = float(input("bar: "))
    j = float(input("pluribus: "))
    k = float(input("sugar: "))
    X=np.array((a,b,c,d,E,f,g,i,j,k))
    Y2=np.dot(X,w)
except Exception as e:
    print (e)
print ("quality of new candy: ",sigmod(Y2))
'''
```
